## Data Visualization

```python
import plotly.express as px
import plotly
```

```python
values=df[' Estimated Unemployment Rate (%)']
category=df['Region']
```

## Pie Charts

```python
fig=px.pie(df , values = values , names=category , title='Estimated Unemployment Rate (%)')
fig.update_traces(textposition="inside" , textinfo="percent+label")
fig.update_layout(autosize=False,
    width=800,
    height=800,
    title_font_size=20
)
```

```python
values=df[' Estimated Employed']
category=df['Region']
```

```python
fig=px.pie(df , values = values , names=category , title='Estimated Employed')
fig.update_traces(textposition="inside" , textinfo="percent+label")
fig.update_layout(autosize=False,
    width=800,
    height=800,
    title_font_size=20
)
```

```python
values=df[' Estimated Labour Participation Rate (%)']
category=df['Region']

fig=px.pie(df , values=values , names=category  , title='Estimated Labour Participation Rate (%)')
fig.update_layout(
    autosize=False,
    width=800,
    height=800
)
```

## Barplots

```python
fig , aixs = plt.subplots(figsize=(10, 10))
plt.subplot(2 ,2  ,1)
sns.barplot(x='Region.1' , y=' Estimated Unemployment Rate (%)' , data=df)
plt.subplot(2 , 2 , 2)
sns.barplot(x='Region.1' , y=' Estimated Employed' , data=df)
plt.subplot(2 , 2 , 3)
sns.barplot(x='Region.1' , y=' Estimated Labour Participation Rate (%)' , data=df)
plt.tight_layout()
```

## Boxplots
